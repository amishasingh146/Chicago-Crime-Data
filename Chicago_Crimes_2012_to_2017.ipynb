# Chicago Crime Dataset for year : 2012 - 2017

Ths is a exploratory data analysis for Chicago Crime Data. The data is taken from Kaggle. 

The data include 1,456,714 rows and following columns : 

__ID & Case number__: These are unique Ids/Primary Keys the crimes.

__Date__: Date of the incident.

__Block__: Block's name where the crime occured.

__IUCR__: Illionis Uniform Crime reporting code is linked to the description of a crime.

__Primary Type__: Primary description of crime according to IUCR.

__Description__: Secondary description of crime according to IUCR.

__Location Description__: Description of location where the incident occured.

__Arrest__: Indicates whether or not an arrest was made.

__Domestic__: Indicates whether the crime was domestic related or not.

__Beat__: Indicates the beat where the incident/crime occured. A beat is a smallest police geographic area.

__District & Ward__: Indicates district and ward where the incident occured.

__Community area__: Indicates community area where the incident occured.

__FBI Code__: Indicates crime classification code as per FBI.

__X and Y co-ordinates__: Indicates X & Y co-ordinates of the location where the incident occured, respectively.

__Year and Updated On__: Indicates time of when the incident occured, and when the incident report was last updated, respectively.

__Latitude, Longitude & Location__: Indicates the latitude and longitude of location where the incident occured.



__Import libraries - Pandas, Numpy, Matplotlib and Seaborn__

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

__Read in Chicago Crime Dataset from Year 2012 to 2017__

```python
chicago = pd.read_csv("Chicago_Crimes_2012_to_2017.csv" , index_col=0)
```

__Show Top 5 rows__

```python
chicago.head()
```

__Statistics of the data__

```python
chicago.describe()
```

__Drop unwanted columns__

```python
chicago_new = chicago.drop(['ID','Case Number','Date','IUCR','FBI Code'],axis=1)
chicago_new.head()
```

# Data Exploration

__Arresets made per year__

```python
Arrests = chicago_new['Arrest'].groupby(chicago_new['Year']).sum()
DataFrame(Arrests)
```

__Domestic Crimes occured per Year__

```python
Domestic = chicago_new['Domestic'].groupby(chicago_new['Year']).sum()
DataFrame(Domestic)
```

__Top 10 Primary Type of Arrests__

```python
Arrests = chicago_new['Arrest'].groupby(chicago_new['Primary Type']).sum()
DataFrame(Arrests.sort_values(ascending = False).head(10))
```

__Top 10 Crime Location__

```python
Arrests = chicago_new['Arrest'].groupby(chicago_new['Location Description']).sum()
DataFrame(Arrests.sort_values(ascending=False).head(10))
```

__Top 10 Districts where Arrests were made__

```python
Arrests = chicago_new['Arrest'].groupby(chicago_new['District']).sum()
DataFrame(Arrests.sort_values(ascending = False).head(10))
```

__Number of times no one was arrested__

```python
len(chicago_new[chicago_new.Arrest != True])
```

# Data Visualization

__Arrests per Year__

```python
ArrestPerYear = chicago_new['Arrest'].groupby(chicago_new['Year']).sum()
ArrestPerYear = DataFrame(ArrestPerYear)
ArrestPerYear = ArrestPerYear.reset_index()

plt.bar(ArrestPerYear['Year'],ArrestPerYear['Arrest'])
plt.show()
```

__Arrests per Primary Type__

```python
ArrestPerType = chicago_new['Arrest'].groupby(chicago_new['Primary Type']).sum()
ArrestPerType = DataFrame(ArrestPerType.sort_values(ascending = False).head(5))
ArrestPerType = ArrestPerType.reset_index()


plt.bar(ArrestPerType['Primary Type'],ArrestPerType['Arrest'])
plt.rcParams['figure.figsize'] = [22, 12]
plt.show()
```
